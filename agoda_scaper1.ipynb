## 利用POST附帶form data的方式爬取網頁，並將網頁內容存在hotel_lists裡

```python
import requests
import time

t1 = time.time()

url = "https://www.agoda.com/api/zh-tw/Main/GetSearchResultList"

hotel_lists = []

for pn in range(6,21):
    pn += 1

    form_data = {
    'SearchMessageID':'36c2cf53-9008-4e8e-89e6-5432b5534a63',
    'IsPollDmc':'false',
    'SearchType':1,
    'ObjectID':0,
    'Filters[PriceRange][Min]':0,
    'Filters[PriceRange][Max]':0,
    'Filters[PriceRange][IsHavePriceFilterQueryParamter]':'false',
    'Filters[ReviewScoreMin]':5,
    'Filters[Size]':0,
    'PlatformID':1001,
    'CurrentDate':'2017-08-02T19:15:04.6961509+07:00',
    'SearchID':991110802191504600,
    'CityId':1784,
    'Latitude':0,
    'Longitude':0,
    'Radius':0,
    'PageNumber':str(pn),
    'PageSize':45,
    'SortType':0,
    'IsSortChanged':'false',
    'SortByAsd':'false',
    'ReviewTravelerType':0,
    'PointsMaxProgramId':0,
    'PollTimes':4,            ## 6至16頁必須將PollTimes更換成4才能順利爬取
    'MaxPollTimes':0,
    'CityName':'京都',
    'ObjectName':'京都',
    'CountryName':'Japan',
    'CountryId':3,
    'IsAllowYesterdaySearch':'false',
    'CultureInfo':'zh-TW',
    'UnavailableHotelId':0,
    'IsEnableAPS':'false',
    'AdditionalExperiments[PRIUS]':1008617,
    'SeletedHotelId':0,
    'HasFilter':'false',
    'LandingParameters[SelectedHotelId]':0,
    'LandingParameters[LandingCityID]':0,
    'NewSSRSearchType':0,
    'IsWysiwyp':'false',
    'FinalPriceView':1,
    'MapType':1,
    'IsShowMobileAppPrice':'false',
    'IsApsPeek':'false',
    'IsRetailPeek':'false',
    'IsRetina':'false',
    'CheckInCultureDateText':'2017/10/7',
    'CheckOutCultureDateText':'2017/10/11',
    'IsCriteriaDatesChanged':'false',
    'CountryEnglishName':'Japan',
    'CityEnglishName':'Kyoto',
    'Adults':2,
    'Children':0,
    'Rooms':1,
    'CheckIn':'2017-10-07T00:00:00',
    'LengthOfStay':4,
    'Text':'京都',
    'IsDateless':'false'    
    }
## 爬取網頁
    res = requests.post(url, data = form_data)
## 經過切字，保留所需的內容       
    hotel = res.text.split('"ResultList":')[1].split(',"SelectedHotel":')[0].replace('【<b class=\\"off-text\\">只限今天</b>】<b class=\\"off-text\\">減價</b><strong class=\\"price-text\\">', '').replace('</strong>","IsContainAGMse":false,"IsAgMse":false,"IsMobilePromotionGrouping":false,"CanTriggerPromotionGrouping":false', '"').replace('\'', "")
## 判斷內容是不是空的，並將內容裝進lists裡     
    if hotel != '[]':
        hotel_lists.append(hotel)

#         time.sleep(3)
t2 = time.time()
print(t2 - t1)
```

```python
## 查詢lists裡的數量和爬取網頁的數是否相同
len(hotel_lists)
```

## 將hotel_lists裡的內容寫成檔案

```python
t1 = time.time()
for j in range(len(hotel_lists)):
    agoda = hotel_lists[j]
    j += 7
    with open('D:/agoda/agoda_20170807_9/agoda_%d.json'%j, encoding='utf-8', mode ='w+') as f:
        f.write(agoda)
t2 = time.time()
print(t2 - t1)
```

## 讀取檔案

```python
import json
import time

count = []

t1 = time.time()

##### 讀取本地端檔案 #####
for file in range(len(hotel_lists)):
    file += 7
    with open('D:/agoda/agoda_20170807_9/agoda_%d.json'%file, encoding='utf-8', mode = 'r+') as f:
        content = f.read() # type = str
        
        json.loads(content)
        json_load = json.loads(content) # type = list
               

# 查詢資料
    try:
        for i in range(len(json_load)):
            json_dict = json_load[i] # type = dict
            print('Name:' + json_dict['TranslatedHotelName'])
            print('Price:' + json_dict['FormattedDisplayPrice'])
            print('StarRating:' + str(json_dict['StarRating']))
            print('Latitude:' + str(json_dict['Latitude']))
            print('Longitude:' + str(json_dict['Longitude']))
            print('Comment:' + str(json_dict['NumberOfReview']))
            print('Score:' + str((json_dict['ReviewScore'])))
            print('=====' * 15)
# 計算旅館數目
            name = json_dict['TranslatedHotelName']
            count.append(name)
    except keyError:
        print('keyError')
        pass
        print('=====' * 15)
f.close()
t2 = time.time()
print('hotel counts:' + ' ' + str(len(count)))
print('spent time:' + ' ' + str(t2-t1))
```
