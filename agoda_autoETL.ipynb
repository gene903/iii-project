## 只要更改"讀取"檔案的檔名&"寫入"檔案的檔名，就可以自動將資料清洗乾淨

```python
import json
import time
import re

hotel_dict = {}
json_str_list = []

t1 = time.time()

##### 讀取本地端檔案 #####
for file in range(0,17):
    file += 1
    with open('D:/agoda/agoda_20170808_24/agoda_%d.json'%file, encoding='utf-8', mode = 'r+') as f:
        content = f.read() # type = str
        
        json.loads(content)
        json_load = json.loads(content) # type = list       

# 從資料中萃取我們要的資訊
    for i in range(len(json_load)): 
        json_dict = json_load[i] # type = dict

        chiName = json_dict['TranslatedHotelName'].replace('?','').split('(')[0] # 切出中文名稱
        latitude = json_dict['Latitude'] # 緯度
        longitude = json_dict['Longitude'] # 經度
        starRate = json_dict['StarRating'] # 飯店星級
        score = json_dict['ReviewScore'] # 旅客評分
        comment = json_dict['NumberOfReview'] # 旅客評論數
        
        price = str(json_dict['FormattedDisplayPrice']) # 飯店價格
        if price == 'None':
            price = 'NA'
        
# 利用正規表示法取出飯店英文名稱        
        string = json_dict['TranslatedHotelName']
    
        engRe = re.findall('(\(\w+)', string) # 正規表示法
      
        if len(engRe) > 0:
            engName = json_dict['TranslatedHotelName'].replace('?','').split('(')[1].split(')')[0]
#             print(EngName)
            hotel_dict["ChiName"]  = chiName.upper()
            hotel_dict["EngName"] = engName.upper()
            hotel_dict["CheckIn"] = "2017-10-07-15-00"
            hotel_dict["CheckOut"] = "2017-10-11-11-00"
            hotel_dict["Price"] = price.replace(',', '')
            hotel_dict["Latitude"] = latitude
            hotel_dict["Longitude"] = longitude
            hotel_dict["StarRate"] = starRate
            hotel_dict["Score"] = score
            hotel_dict["Comment"] = comment
        else:
            hotel_dict["ChiName"]  = chiName.upper()
            hotel_dict["EngName"] = chiName.upper()
            hotel_dict["CheckIn"] = "2017-10-07-15-00"
            hotel_dict["CheckOut"] = "2017-10-11-11-00"
            hotel_dict["Price"] = price.replace(',', '')
            hotel_dict["Latitude"] = latitude
            hotel_dict["Longitude"] = longitude
            hotel_dict["StarRate"] = starRate
            hotel_dict["Score"] = score
            hotel_dict["Comment"] = comment
#         print(hotel_dict)
        json_str = json.dumps(hotel_dict, ensure_ascii=False) # 將 python的 dict格式轉換成 json格式，type = str
#         print(json_str)
        json_str_list.append(json_str) # 將json字串塞入一個 list
    print(json_str_list)
# 將 python的 list格式轉換成 json格式，type = str   
    agoda_ETL = json.dumps(json_str_list, ensure_ascii=False).replace('\\', '').replace('"{', '{').replace('}"', '}').replace('\xa0', '')
# 將處理好的資訊寫進檔案中    
    with open('D:/agoda/ETL/agoda_20170808_25.json', encoding='utf-8', mode = 'w+') as f:
        f.write(agoda_ETL)
    
f.close()
t2 = time.time()
print('spent time:' + ' ' + str(t2-t1))
```
