## 產生URL

```python
import requests
import time

t1 = time.time()

url_lists = []
content_lists = []
hotel_lists = []
count = [] 

## 使用list包住所有url
pg = 0
checkin = "2017-10-07"
checkout = "2017-10-11"
while pg < 5:
    pg += 1   
    for pn in range(1,50):       
        url = "https://tw.hotels.com/search/listings.json?destination-id=725674&q-destination=%E4%BA%AC%E9%83%BD,+%E6%97%A5%E6%9C%AC&sort-order=STAR_RATING_HIGHEST_FIRST&q-check-in="+ checkin + "&q-check-out="+ checkout + "&pg="+str(pg)+"&pn="+str(pn)
        url_lists.append(url)

print(len(url_lists))

t2 = time.time()
print(t2 - t1)
```

## 將網頁內容存在hotel_lists裡

```python
t1 = time.time()

for i in range(len(url_lists)):
    res = requests.get(url_lists[i])
    content = res.text.split('"results":')[1].split(',"debugInfo":')[0].split(',"unavailableCount"')[0]
    if content != '[]':
        content_lists.append(content)

for j in range(len(content_lists)):
    if content_lists[j] != content_lists[j-1]:
        hotel_lists.append(content_lists[j])
#         print(content_lists[j])

print('file counts' + ' ' + str(len(hotel_lists)))

t2 = time.time()
print(t2 - t1)
```
